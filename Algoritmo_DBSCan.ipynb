# Algoritmo DBScan

### Viene de las palabras Density Based Spatial Clustering of Applications. Es un algoritmo NO supervisado presentado en 1996 por Martin Ester, Hans-Peter Kriegel, JÃ¶rg Sander y Xiawei Xu.

### El agoritmo divide un conjunto de puntos en tres subconjuntos:
#### Puntos core: Cumplen con las condiciones de densidad fijadas.
#### Puntos alcanzables: NO cumplen con las condiciones de densidad fijadas pero tienen cerca otros puntos core.
#### Ruido: NO son puntos core ni alcanzables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans, DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial import distance
from sklearn.neighbors import NearestNeighbors
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=500, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.title("DBSCAN con eps=0.3, min_samples=5")
plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=1000, centers=5, cluster_std=1.6, random_state=42)

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.title("DBSCAN con eps=0.5, min_samples=5")
plt.show()
```

```python
dbscan2 = DBSCAN(eps=0.3, min_samples=10)
dbscan2.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=dbscan2.labels_)
plt.title("DBSCAN con eps=0.3, min_samples=10")
plt.show()
```
