# Mathematical Definition of a Pandas Series

A **Pandas Series** can be defined as a one-dimensional array-like structure, indexed and labeled, represented as:

$$
S = \begin{pmatrix}
s_1 \\
s_2 \\
\vdots \\
s_n \\
\end{pmatrix}
$$

where $S$ is the Series, $n$ is the number of elements, and each $s_i$ represents the value at position $i$. Additionally:


-     The elements are indexed by a set of indices $I = \{i_1, i_2, \dots, i_n\}$.
-     Each element $s_i$ can belong to any data type (e.g., numeric, string, boolean).


**Vector Interpretation:** If all elements of the Series are numeric, it can be viewed as a vector in an $n$-dimensional space:

$$
\mathbf{S} =
\begin{pmatrix}
s_1 \\
s_2 \\
\vdots \\
s_n \\
\end{pmatrix}
\quad \text{where } \mathbf{S} \in \mathbb{R}^n
$$

In this context, the Series $S$ is a single column vector in the vector space $\mathbb{R}^n$.



```python
%%time

# Creating a Series from a list
data = [10, 20, 30, 40, 50]      # Creates a list named 'data' with five integer values.
series = pd.Series(data)         # Converts the 'data' list into a Pandas Series, which is like a column of data in a table.
print(series)                    # Prints the Pandas Series to the console, displaying the data and their corresponding indices.

```

# Mathematical Definition of a DataFrame

A **DataFrame** can be defined as a two-dimensional matrix with indices and labels, represented as:


$$DF =
\begin{pmatrix}
x_{11} & x_{12} & \cdots & x_{1n} \\
x_{21} & x_{22} & \cdots & x_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
x_{m1} & x_{m2} & \cdots & x_{mn} \\
\end{pmatrix}
$$

where \(DF\) is the DataFrame, \(m\) is the number of rows, \(n\) is the number of columns, and each $(x_{ij})$ represents the value in row \(i\) and column \(j\). Additionally:

- The rows are indexed by a set of indices $$I = \{i_1, i_2, \dots, i_m\}$$
- The columns are labeled by a set of labels $$C = \{c_1, c_2, \dots, c_n\}$$

**Columns as Vectors:** Each column of the DataFrame can be viewed as a vector in an \(m\)-dimensional space:

$$
\mathbf{c_j} =
\begin{pmatrix}
x_{1j} \\
x_{2j} \\
\vdots \\
x_{mj} \\
\end{pmatrix}
\quad \text{for } j = 1, 2, \dots, n
$$

**Vector Space:** If all columns of the DataFrame are numeric, the rows can be interpreted as points in an \(n\)-dimensional vector space $\mathbb{R}^n$.
   The entire DataFrame can be considered as a set of \(m\) points in $\mathbb{R}^n$

```python
%%time
# Creating a DataFrame from a dictionary
dd = {'Name': ['John', 'Anna', 'Peter'],  # Creates a dictionary with keys 'Name', 'Age', and 'City', each associated with a list of values
      'Age': [28, 24, 22],               # 'Name' has a list of names, 'Age' has a list of ages
      'City': ['New York', 'Los Angeles', 'Chicago']}  # 'City' has a list of cities

df1 = pd.DataFrame(dd)  # Converts the dictionary 'dd' into a DataFrame (a table-like data structure in pandas)

df1  # Displays the DataFrame, showing the 'Name', 'Age', and 'City' columns with their respective values

```

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">7.1) Aggregations (sum, mean, etc.): <h2/>

```python
%%time
# Assuming df_csv is already defined and loaded with your data

# Mean
media_num_athletes = df_csv['num_athletes'].mean()  # Calculate the average (mean) of the 'num_athletes' column
media_num_coaches = df_csv['num_coaches'].mean()  # Calculate the average (mean) of the 'num_coaches' column

print(f"Media de num_athletes: {media_num_athletes}")  # Print the mean of the 'num_athletes' column
print(f"Media de num_coaches: {media_num_coaches}")  # Print the mean of the 'num_coaches' column

# Median
mediana_num_athletes = df_csv['num_athletes'].median()  # Calculate the median of the 'num_athletes' column
mediana_num_coaches = df_csv['num_coaches'].median()  # Calculate the median of the 'num_coaches' column

print(f"Mediana de num_athletes: {mediana_num_athletes}")  # Print the median of the 'num_athletes' column
print(f"Mediana de num_coaches: {mediana_num_coaches}")  # Print the median of the 'num_coaches' column

# Mode
moda_num_athletes = df_csv['num_athletes'].mode()[0]  # Calculate the mode (most frequent value) of the 'num_athletes' column and take the first value
moda_num_coaches = df_csv['num_coaches'].mode()[0]  # Calculate the mode (most frequent value) of the 'num_coaches' column and take the first value

print(f"Moda de num_athletes: {moda_num_athletes}")  # Print the mode of the 'num_athletes' column
print(f"Moda de num_coaches: {moda_num_coaches}")  # Print the mode of the 'num_coaches' column

```

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">7.2) Groupings (groupby):</h2>

```python
%%time
# Grouping by country and calculating the average number of athletes and coaches
promedio_por_pais = df_csv.groupby('country')[['num_athletes', 'num_coaches']].mean().reset_index()  # Groups the data by the 'country' column and calculates the mean of 'num_athletes' and 'num_coaches' for each country, then resets the index to turn the result back into a DataFrame.
print(promedio_por_pais)  # Prints the DataFrame showing the average number of athletes and coaches per country.

# Finding the event with the most participants
evento_mas_participantes = df_csv.groupby('events')['athletes'].sum().idxmax()  # Groups the data by the 'events' column, sums up the 'athletes' column for each event, and then finds the event with the maximum number of participants.
print(f"The event with the most participants is: {evento_mas_participantes}")  # Prints the event with the highest number of participants.
```

```python
%%time

# Suppose df_csv is your DataFrame
# First, ensure that 'num_athletes' and 'num_coaches' columns are numeric
df_csv['num_athletes'] = pd.to_numeric(df_csv['num_athletes'], errors='coerce')  # Convert 'num_athletes' column to numeric, invalid parsing will be set as NaN
df_csv['num_coaches'] = pd.to_numeric(df_csv['num_coaches'], errors='coerce')  # Convert 'num_coaches' column to numeric, invalid parsing will be set as NaN

# Next, calculate the proportion of athletes per coach
df_csv['athletes_per_coach'] = df_csv['num_athletes'] / df_csv['num_coaches']  # Divide the number of athletes by the number of coaches to get the ratio

# Handle cases where num_coaches might be 0 to avoid division by zero errors
df_csv['athletes_per_coach'] = df_csv['athletes_per_coach'].replace([np.inf, -np.inf], np.nan)  # Replace infinite values with NaN

# Now, create a new column 'management_status' using a lambda function
df_csv['management_status'] = df_csv['athletes_per_coach'].apply(lambda x: 'Well managed' if x <= 10 else 'Poorly managed')  # Apply a function that labels management status based on the ratio of athletes per coach

# Print the DataFrame to see the new column
print(df_csv[['country', 'num_athletes', 'num_coaches', 'athletes_per_coach', 'management_status']].head())  # Display the first few rows of the DataFrame with selected columns
```


<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">8.3) Rename columns:</h2>

```python
%%time
# Assuming df_csv is your DataFrame
# First, ensure that the columns num_athletes and num_coaches are numeric
df_csv['num_athletes'] = pd.to_numeric(df_csv['num_athletes'], errors='coerce')
df_csv['num_coaches'] = pd.to_numeric(df_csv['num_coaches'], errors='coerce')

# Then, calculate the ratio of athletes per coach
df_csv['athletes_per_coach'] = df_csv['num_athletes'] / df_csv['num_coaches']

# Handle cases where num_coaches could be 0 to avoid division by zero
df_csv['athletes_per_coach'] = df_csv['athletes_per_coach'].replace([np.inf, -np.inf], np.nan)

# Now, create a new column 'management_status' using a lambda function
df_csv['management_status'] = df_csv['athletes_per_coach'].apply(lambda x: 'Well Managed' if x <= 10 else 'Poorly Managed')

# Renaming columns for better readability
df_csv.rename(columns={
    'code': 'team_code',
    'team': 'team_name',
    'team_gender': 'team_gender',
    'country': 'country_short',
    'country_full': 'country_full_name',
    'country_code': 'country_code',
    'discipline': 'discipline',
    'disciplines_code': 'discipline_code',
    'events': 'events',
    'athletes': 'athletes',
    'coaches': 'coaches',
    'athletes_codes': 'athletes_codes',
    'num_athletes': 'number_of_athletes',
    'coaches_codes': 'coaches_codes',
    'num_coaches': 'number_of_coaches',
    'athletes_per_coach': 'athletes_per_coach_ratio',
    'management_status': 'management_status'
}, inplace=True)

# Print the DataFrame to see the new column names
print(f'The df has the renewed columns as: {df_csv}')
```

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">9.1) Merge :</h2>


```python
%%time


# Sample DataFrame for teams
data_teams = {
    'team_code': ['T1', 'T2', 'T3'],
    'team_name': ['Team A', 'Team B', 'Team C'],
    'country_short': ['USA', 'CAN', 'AUS'],
    'number_of_athletes': [10, 15, 20],
    'number_of_coaches': [2, 3, 4]
}
df_teams = pd.DataFrame(data_teams)

print(f'df_teams >>>> \n{df_teams}')

# Sample DataFrame for countries
data_countries = {
    'country_short': ['USA', 'CAN', 'AUS', 'GBR'],
    'country_full_name': ['United States', 'Canada', 'Australia', 'United Kingdom'],
    'continent': ['North America', 'North America', 'Australia', 'Europe']
}
df_countries = pd.DataFrame(data_countries)

print(f'df_teams >>>> \n{df_countries}')
# Perform a merge operation based on the 'country_short' column
df_merged = pd.merge(df_teams, df_countries, on='country_short')

# Print the merged DataFrame
print(f'df_merged  >>>> \n{df_merged}')
```


<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">9.2) Concat :</h2>

```python
%%time

# Sample DataFrame 1
data1 = {
    'team_code': ['T1', 'T2', 'T3'],
    'team_name': ['Team A', 'Team B', 'Team C'],
    'country_short': ['USA', 'CAN', 'AUS'],
    'number_of_athletes': [10, 15, 20],
    'number_of_coaches': [2, 3, 4]
}
df_a = pd.DataFrame(data1)
print(f'df_a  >>>> \n{df_a}')
# Sample DataFrame 2
data2 = {
    'team_code': ['T4', 'T5', 'T6'],
    'team_name': ['Team D', 'Team E', 'Team F'],
    'country_short': ['GBR', 'FRA', 'GER'],
    'number_of_athletes': [12, 18, 22],
    'number_of_coaches': [2, 3, 4]
}
df_b = pd.DataFrame(data2)
print(f'df_b  >>>> \n{df_b}')
# Row-wise concatenation
df_concat_rows = pd.concat([df_a, df_b], ignore_index=True)

# Print the concatenated DataFrame
print("<<<<<<<<<< Row-wise concatenation: >>>>>>>>>>>> \n")
print(df_concat_rows)
```

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;">12.1) How to filter words that contain atleast 2 vowels from a series?</h2>


```python
%%time
# input
ser = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money'])

'''
Desired Output


0     Apple
1    Orange
4     Money
dtype: object
'''

# using nested loops
vowels = list("aeiou")
list_ = []
for w in ser:
    c = 0
    for l in list(w.lower()):
        if l in vowels:
            c += 1
    if c >= 2:
        print(w)
        list_.append(w)

ser[ser.isin(list_)]

# another solution using counter

from collections import Counter
mask = ser.map(lambda x: sum([Counter(x.lower()).get(i, 0) for i in list('aeiou')]) >= 2)
ser[mask]
```
